```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from IPython.core.display import display, HTML
#from apyori import apriori  

```

```python
## Outra forma de aplicação das regras de associação
dados_set_2017 = pd.read_csv('mp25_matriz_Setembro_2017.csv',sep=';',encoding = "ISO-8859-1")

dados_set_2017
```

```python
def force_show_all(df):
    with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.width', None):
        display(HTML(df.to_html()))

pd.options.display.max_rows = 999

frequencia_itens_2017 = apriori(dados_set_2017, min_support=0.80, use_colnames=True)

regras = association_rules(frequencia_itens_2017)
force_show_all(regras)


```
