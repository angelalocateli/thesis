```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
#from apyori import apriori  

```

```python
## Outra forma de aplicação das regras de associação
dados_set_2017 = pd.read_csv('matiz_Setembro_2017_csv.csv',sep=';',encoding = "ISO-8859-1")

dados_set_2017
```

```python


frequencia_itens_2017 = apriori(dados_set_2017, min_support=0.70, use_colnames=True)

regras = association_rules(frequencia_itens_2017)
regras



```
